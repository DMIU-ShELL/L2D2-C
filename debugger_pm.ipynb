```python
import pickle
import numpy as np
import torch
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def load_data(fpath, num_tasks):
    precision_matrices = []
    for task_idx in range(1, num_tasks+1):
        with open(fpath.format(task_idx), 'rb') as f: precision_matrices.append(pickle.load(f))
    return precision_matrices
```

```python
agent =  'A2C'
pm_method = 'MAS'
env = 'CTgraph-v0'
num_tasks = 3
path = './log/CTgraph-v0-a2c-mas-pres_actor_and_critic_balnet-sep-pres/210909-225614/'
```

### Per task precision matrices

```python
# precision matrices after executing consolidation method at end of task
fpath = path + '/' + '{0}Agent{1}-vanilla-precision-matrices-{2}'.format(agent, pm_method, env)
fpath = fpath + '-task-{0}.bin'
precision_matrices = load_data(fpath, num_tasks)
```

```python
for i, pm in enumerate(precision_matrices):
    print('-----> pm, task {0}'.format(i+1))
    for k, v in pm.items():
        v = v.detach().cpu().numpy()
        print('key:', k, '/', 'value.shape:', v.shape)
        print('value.max():', v.max())
        print('value.min():', v.min())
        v = v.reshape(-1)
        print('percentage of unused params: {0:.2f}%'.format((v==0.).sum() / len(v)*100))
        plt.hist(np.log10((1e+4)*v+1e-12),bins=100)
        #plt.hist(v,bins=100)
        plt.show()
```
