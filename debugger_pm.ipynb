```python
import json
import pickle
import numpy as np
import torch
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
# def load_data(fpath, num_tasks):
#     precision_matrices = []
#     for task_idx in range(1, num_tasks+1):
#         with open(fpath.format(task_idx), 'rb') as f: precision_matrices.append(pickle.load(f))
#     return precision_matrices

def load_data(fpath, num_runs, num_tasks):
    precision_matrices = []
    for run_idx in range(1, num_runs+1):
        for task_idx in range(1, num_tasks+1):
            with open(fpath.format(run_idx, task_idx), 'rb') as f: precision_matrices.append(pickle.load(f))
    return precision_matrices
```

```python
#path = './log/CTgraph-v0-ppo-scp-kwinners-5percent/211220-051117/'
path = './log/CTgraph-v0-ppo-scp-sparse-group-l1-weights/211218-034602/'
with open('{0}/config.json'.format(path), 'r') as f: 
    config = json.load(f)
num_tasks = config['cl_num_tasks']
num_learning_blocks = config['cl_num_learn_blocks'] # number of times/runs all tasks are repeatedly trained on.
```

### Per task precision matrices

```python
# precision matrices after executing consolidation method at end of task
fpath = path + '/pm/' + '{0}-vanilla-precision-matrices-{1}'.format(config['agent_name'], config['env_name'])
fpath = fpath + '-run-{0}-task-{1}.bin'
precision_matrices = load_data(fpath, num_learning_blocks, num_tasks)
```

```python
stats_used_params = {}
stats_used_params['tasks_idx'] = []

for i, pm in enumerate(precision_matrices):
    stats_used_params['tasks_idx'].append('task {0}'.format(i % num_tasks +1))
    print('-----> pm, task {0}'.format(i % num_tasks +1))
    for k, v in pm.items():
        v = v.detach().cpu().numpy()
        print('key:', k, '/', 'value.shape:', v.shape)
        print('value.max():', v.max())
        print('value.min():', v.min())
        v = v.reshape(-1)
        used_params = ((v!=0.).sum() / len(v)) *100 # percentage of used parameters
        print('percentage of used params: {0:.2f}%'.format(used_params))
        plt.hist(np.log10((1e+4)*v+1e-12),bins=100)
        #plt.hist(v,bins=100)
        plt.show()
        
        if k not in stats_used_params.keys(): stats_used_params[k] = []
        stats_used_params[k].append(used_params)
```

```python
df_stats_used_params = pd.DataFrame(stats_used_params)
df_stats_used_params.T
```
